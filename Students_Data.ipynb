```python
Students=['nani','srinivas','yogitha']
CollegeFee=[45000,56000,78000]
Marks=[43,67,60]
print(Students)
print(CollegeFee)
print(Marks)
print(len(Students))
print(len(CollegeFee))
print(len(Marks))
#display available data

def display_data():
  for i in range(len(Students)):
    print(f"{Students[i]} - CollegeFee : {CollegeFee[i]} - Marks : {Marks[i]}")
display_data()

#updating CollegeFee

def update_CollegeFee(StudentName,new_fee):
  if StudentName in Students:
    index = Students.index(StudentName)
    CollegeFee[index]=new_fee
    print(f"updated {StudentName} fee to ${new_fee}")
  else:
    print("StudentName not found!")
update_CollegeFee("srinivas",50000)

#updating marks

def update_Marks(StudentName,new_marks):
  if StudentName in Students:
        index = Students.index(StudentName)
        Marks[index]=new_marks
        print(f"updated {StudentName} Marks to {new_marks}")
  else:
        print("StudentName not found!")
update_Marks("nani",25)
#search students
def search_Students(StudentName):
  if StudentName in Students:
    index = Students.index(StudentName)
    print(f"{StudentName} - CollegeFee:${CollegeFee[index]}, Marks:{Marks[index]}")
  else:
    print("StudentName not found!")
search_Students("yogitha")






```
